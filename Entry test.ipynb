```python
# Câu 2: Dùng thư viện opencv load ảnh image.png và thực hiện các bước sau:

# Cắt góc phần tư trái trên của ảnh
# Resize ảnh, dài rộng còn một nửa
# Thực hiện Gaussian blur ảnh
# Phát hiện edge trong ảnh
```

```python
import cv2
from matplotlib import pyplot as plt

# Hiển thị ảnh
img = cv2.imread("image.png")
plt.imshow(img)
plt.show()
```

```python
# Lấy kích thước của ảnh

img = cv2.imread("image.png")
(h, w, d) = img.shape
print("height={}, width={}, depth={}".format(h, w, d))
```

```python
# Cắt góc phần tư trái trên của ảnh

roi = img[0:h//2, 0:w//2]
cv2.imshow('Region Of Interest', roi)
cv2.waitKey(0)
```

```python
# Resize ảnh, dài rộng còn một nửa

import cv2
 
img = cv2.imread("image.png")
print('Original Dimensions : ',img.shape)
 
scale_percent = 50  # percent of original size
height = int(img.shape[0]*scale_percent/100)
width = int(img.shape[1]*scale_percent/100)
dim = (width, height)
  
# resize image
resized = cv2.resize(img, dim)
print('Resized Dimensions : ',resized.shape)
 
cv2.imshow("Resized image", resized)
cv2.waitKey(0)
```

```python
# Thực hiện Gaussian blur ảnh

import cv2
 
img = cv2.imread("image.png")
blured = cv2.GaussianBlur(img, (3,3), 1)

cv2.imshow("Blured image", blured)
cv2.waitKey(0)
```

```python
# Phát hiện edge trong ảnh

import cv2
 
img = cv2.imread("image.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edged = cv2.Canny(gray, 50, 100)

cv2.imshow("Edged image", edged)
cv2.waitKey(0)
```
